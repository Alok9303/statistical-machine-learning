```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
df.columns
```

```python
dir(df)
```

```python
df=df.rename(columns={'per capita income (US$)':'income'})
```

```python
df
```

```python
df.dtypes
```

```python
%matplotlib inline
plt.scatter(df.year,df.income,color='red',marker='+')
plt.xlabel('year')
plt.ylabel('income((US$)')
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year']],df.income)   #also I can give df.price as df['price']
```

Predict the income in the year 2025

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d=pd.read_csv('to-predict-years.csv')
```

```python
d
```

```python
p=reg.predict(d)
```

```python
d['predicted_income']=p
```

```python
d
```

```python
d.to_csv('predicted-income.csv',index=False)
```

Original income vs predicted income

```python
plt.scatter(df.year,df.income,color='red',marker='+')
plt.xlabel('year')
plt.ylabel('income((US$)')
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
```

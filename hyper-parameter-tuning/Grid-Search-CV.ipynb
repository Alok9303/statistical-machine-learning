```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
d = load_iris()
```

```python
dir(d)
```

```python
d.target_names
```

```python
df = pd.DataFrame(d.data,columns = d.feature_names)   #create a dataframe
```

```python
df.head()
```

```python
df['flower'] = d.target  #create column
```

```python
df.head()
```

```python
df['flower_name'] = df['flower'].apply(lambda x : d.target_names[x])
```

```python
df.head()
```

```python
X = df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
y = df.flower
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(SVC(gamma = 'auto'),{'kernel':['rbf','linear','poly'],'C':[1,2,5,10]},cv = 5, return_train_score=False)
```

```python
clf.fit(X,y)
```

```python
clf_df = pd.DataFrame(clf.cv_results_)
```

```python
clf_df.head()
```

```python
clf_df[['params','mean_test_score']]
```

```python
dir(clf)
```

```python
clf.best_params_   #these are the optimal parameters.
```

```python
clf.best_score_
```

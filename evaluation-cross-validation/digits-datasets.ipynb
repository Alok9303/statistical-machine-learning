```python
from sklearn.datasets import load_digits
```

```python
d = load_digits()
```

```python
dir(d)
```

```python
X = d.data
y = d.target
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
cv_lr = cross_val_score(LogisticRegression(max_iter = 1000),X,y)
```

```python
cv_svm = cross_val_score(SVC(),X,y)
```

```python
cv_rf = cross_val_score(RandomForestClassifier(n_estimators = 500),X,y)
```

```python
print(cv_lr.mean())
```

```python
print(cv_svm.mean())
```

```python
print(cv_rf.mean())
```

since average score of svm is highest, it is the best classifier for this perticular dataset.

Create a Dataframe

```python
import pandas as pd
```

```python
df = pd.DataFrame(d.data, columns = d.feature_names)
```

```python
df.head()
```

```python
X = df
y = d.target
```

```python
X.head()
```

```python
X.describe()   # I got to know that min=0 and max=16. So we will scale it to standards.
```

```python
import numpy as np
```

```python
np.unique(y)   # Properly scaled. We have to scale X.
```

We have to scale X.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X = X_scaled  #just saving it as X for convinience.
```

```python
X = pd.DataFrame(X,columns = d.feature_names)    #just saving it again as a Dataframe.
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

Dimensionality Reduction using PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(0.95)   #means the new components holds 95% of the information(variance).
```

```python
dir(pca)
```

```python
X_pca = pca.fit_transform(X)
```

```python
X_pca.shape    #no. of components after PCA with 95% is 40.
```

```python
# We can also check the variance of each components.(the info each component is providing.)
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
#lets train the model now using X_pca.
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train_pca,y_train)
```

```python
model.score(X_test_pca,y_test)
```

Even though I have done dimensionality reduction there is no significant change in the accuracy.

Other way, using the n_components parameter.

```python
pca = PCA(n_components = 2)
```

```python
X_pca = pca.fit_transform(X)
X_pca.shape  
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train_pca,y_train)
```

```python
model.score(X_test_pca,y_test)
```

As I have reduced the dimension to 2 all the way from 64, It had lost so much of information and hence the accuracy is less.

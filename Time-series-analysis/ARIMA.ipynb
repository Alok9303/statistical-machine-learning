```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
data = [10,12,14,16,18,20,20,22] #linear data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data)
```

```python
from statsmodels.tsa.arima.model import ARIMA  # not suitable when there is seasonality.
```

```python
model = ARIMA(data, order = (1,1,1))   # (p,d,q)=> (AR lags,differencing,MA lags)
```

```python
fitted_model = model.fit()
```

```python
y_pred = fitted_model.predict(8,10).tolist()
```

```python
combined = data + y_pred
```

```python
plt.plot(combined)
```

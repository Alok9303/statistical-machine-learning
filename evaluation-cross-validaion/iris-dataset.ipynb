```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
cv_lr = cross_val_score(LogisticRegression(max_iter = 150),iris.data,iris.target)
cv_svm = cross_val_score(SVC(),iris.data,iris.target)
cv_rf = cross_val_score(RandomForestClassifier(n_estimators=150),iris.data,iris.target)
```

```python
print(cv_lr.mean())
```

```python
print(cv_svm.mean())
```

```python
print(cv_rf.mean())
```

All the models gave the same results almost. But Linear Regression gave the highest.

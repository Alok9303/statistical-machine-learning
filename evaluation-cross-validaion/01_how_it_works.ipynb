```python
data = [1,2,3,4,5,6,7,8,9]
y = ['odd','even','odd','even','odd','even','odd','even','odd']
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
df['type'] = y
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['target'] = le.fit_transform(df.type)
```

```python
df
```

```python
X = df.drop(['type','target'], axis = 'columns')
y = df.target
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits = 3)
```

```python
for i in kf.split(X):
    print(i)
```

```python
for i,j in kf.split(X):
    print(i,j)
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits = 3)
```

```python
for i,j in skf.split(X,y):
    print(i,j)
```

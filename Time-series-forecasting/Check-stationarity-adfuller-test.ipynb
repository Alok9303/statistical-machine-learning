## To check Stationarity by Augmented Dickey-Fuller test(adfuller test)

In this test , the null hypothesis is time series data is non-stationary.

Remember:: adfuller(data)[1]  gives the p-values

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adf_test = adfuller(data)
```

```python
adf_test[1]
```

p-value = 0.9 > 0.05  means we accept null hypothesis. So ts data is not stationary.

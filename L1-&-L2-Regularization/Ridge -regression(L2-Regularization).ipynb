```python
import pandas as pd
```

```python
dataset = pd.read_csv('Melbourne_housing_FULL.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.nunique()
```

```python
cols_to_use = ['Suburb','Rooms','Type','Price','Method','SellerG','Distance','Bedroom2','Bathroom','Car','Landsize','BuildingArea','CouncilArea','Regionname','Propertycount']
```

```python
dataset = dataset[cols_to_use]
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Bedroom2','Bathroom','Car','Distance','Propertycount']
```

```python
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_mean = ['Landsize','BuildingArea']
```

```python
dataset[cols_to_fill_mean] = dataset[cols_to_fill_mean].fillna(dataset[cols_to_fill_mean].mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace = True)
```

```python
dataset.isna().sum()
```

```python
##We have to convert categorical data into numerical data 
```

```python
dataset = pd.get_dummies(dataset,dtype=int, drop_first = True)
```

```python
dataset
```

```python
X = dataset.drop('Price', axis = 1)
y = dataset.Price
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state=2)
```

```python
##Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg_model = LinearRegression()
```

```python
reg_model.fit(X_train,y_train)
```

```python
reg_model.score(X_test,y_test)
```

```python
reg_model.score(X_train,y_train)
```

So from the above we can say that the model is accurate only for the train data and not performing good for the test data. i,e the model is overfitting

So we can use regularization to reduce the overfitting

```python
from sklearn.linear_model import Ridge
```

```python
ridge_reg_model = Ridge(alpha = 50, max_iter = 500, tol = 10)
```

```python
ridge_reg_model.fit(X_train,y_train)
```

```python
ridge_reg_model.score(X_test,y_test)
```

```python
ridge_reg_model.score(X_train,y_train)
```

So, now after the L2 regularization, we reduced the overfitting. and now the model is balanced fit.

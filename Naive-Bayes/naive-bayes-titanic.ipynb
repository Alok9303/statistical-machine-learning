```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

Data Exploration.

```python
df.head(4)
```

Sepearte inputs and outputs.

```python
inputs = df.drop(['PassengerId','Pclass','Survived','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis = 'columns')
inputs.head(4)
```

outputs.

```python
y = df.Survived
y.head(4)
```

Convert the columns with categories to numerical data.

```python
df_dummies = pd.get_dummies(inputs['Sex'], dtype = 'int')
df_dummies
```

Merge the dummies dataframe with the original input dataframe.

```python
merged = pd.concat([inputs,df_dummies], axis = 'columns')
```

```python
merged.head(4)
```

Drop the categorical column.

```python
X = merged.drop('Sex', axis = 'columns')
```

```python
X
```

Remove null values by replacing na value with the mean.

```python
#we have to check for na values.
```

```python
X.isnull().any()
```

```python
#OR
X.columns[X.isna().any()]
```

```python
X.Age = X['Age'].fillna(X.Age.mean())
```

```python
X.Age.head(6)   #6th value is filled with the mean value.(29.699118)
```

Split the data into training and testing datasets.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

import and create the naive bayes model

GaussianNB is used for normal distribution.

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
#Training of the model.
model.fit(X_train,y_train)
```

Accuracy of the model.

```python
model.score(X_test,y_test)
```
